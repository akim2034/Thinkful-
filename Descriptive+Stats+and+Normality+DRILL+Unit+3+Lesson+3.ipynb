```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Binomial Distribution

```python
binomial = np.random.binomial(20, .5, 100)

plt.hist(binomial, alpha=.7)

plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean()-binomial.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Binomial Distribution')

plt.show()
```

The mean and std gives us an idea about where most of the values lie. However, the graph can look more scattered in some cases, so I am not yet positive if the mean and std actually helps shed insight about the graph. 

## 2. Chisquare Distribution

```python
chisquare = np.random.chisquare(2, 100)

plt.hist(chisquare, alpha=.7)

plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chisquare.mean()- chisquare.std(), color='b', linestyle='dashed', linewidth=2) 
plt.title('Chisquare Distribution')

plt.show()
```

The mean and std does not tell us much about the graph. The outliers will tell us more about the shape of the curve being skewed to the right.

For gamma distribution, it also seems that the mean and std does not tell us much about the shape of the curve. It is skewed to the right.
